```python
import pickle
import pandas as pd
import numpy as np
```

```python
FILE_PATH = "/content/drive/MyDrive"
data = pd.read_csv(f"{FILE_PATH}/preprocessed_tweets.csv")
data.head()
```

```python
unique_dates = data['date'].unique()  # Get unique values in the 'date' column
count_unique_dates = len(unique_dates)  # Count the number of unique dates

print("Unique dates:", unique_dates)
print("Count of unique dates:", count_unique_dates)
```

```python
with open(f'{FILE_PATH}/vectorizer.pkl', 'rb') as file:
    loaded_vectorizer = pickle.load(file)

# Load the trained model
with open(f'{FILE_PATH}/calibrated_svc_model.pkl', 'rb') as file:
    loaded_model = pickle.load(file)
```

```python
X = data['text_without_stopwords']
vecX = loaded_vectorizer.transform(X)

y_pred_proba = loaded_model.predict_proba(vecX)

y_pred_proba
```

```python
probabilities_df = pd.DataFrame(y_pred_proba, columns=loaded_model.classes_)

data_with_probabilities = pd.concat([data, probabilities_df], axis=1)
```

```python
data_with_probabilities.head()
```

```python
data_with_probabilities.to_csv("tweets_proba.csv")
```

```python
# Group by date and calculate the average sentiment probabilities
daily_sentiment = data_with_probabilities.groupby('date')[[0, 1]].mean()

# Add count of tweets for each day
daily_sentiment['volume'] = data_with_probabilities.groupby('date').size()

print(daily_sentiment)
```

```python
daily_sentiment.to_csv("daily_sentiment.csv")
```
